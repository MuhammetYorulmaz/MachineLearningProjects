# Bank Marketing DataSet - Banka Pazarlama Veri Seti

__Pazarlama Tanımı:__ Şirketlerin müşteriler için değer yaratma ve karşılığında müşterilerden değer elde etmek için güçlü müşteri ilişkileri kurma süreci.

__Pazarlama kampanyaları__, müşteri ihtiyaçlarına ve genel memnuniyetlerine odaklanarak karakterize edilir. Bununla birlikte, bir pazarlama kampanyasının başarılı olup olmayacağını belirleyen farklı değişkenler vardır. Pazarlama kampanyası yaparken göz önünde bulundurmamız gereken bazı değişkenler vardır.
* __Nüfus Segmenti:__ Pazarlama kampanyası nüfusun hangi kesimine hitap edecek ve neden? Pazarlama kampanyasının bu yönü, pazarlama kampanyasının mesajını büyük olasılıkla nüfusun hangi bölümünün alması gerektiğini söyleyeceğinden son derece önemlidir.
* __Müşterinin bulunduğu yere ulaşmak için dağıtım kanalı:__ Nüfusun hangi kesimine hitap etmeliyiz? Mesajımızı iletmek için hangi aracı kullanmalıyız? (Örn: Telefonlar, Radyo, TV, Sosyal Medya vb.)
* __Fiyat:__ Potansiyel müşterilere teklif edilecek en iyi fiyat nedir? (Bankanın pazarlama kampanyası söz konusu olduğunda, bu gerekli değildir, çünkü bankanın asıl ilgisi potansiyel müşterilerin bankanın operasyonel faaliyetlerinin devam etmesini sağlamak için mevduat hesabı açmasıdır.)
* __Promosyon Stratejisi:__ Bu, stratejinin uygulanma şekli ve potansiyel müşterilere nasıl hitap edileceğidir. Bu, pazarlama kampanyası analizinin son kısmı olmalıdır, çünkü önceki hatalardan ders çıkarmak ve pazarlama kampanyasının nasıl daha etkili hale getirileceğini belirlemek için (mümkünse) önceki kampanyaların derinlemesine bir analizi yapılmalıdır.

**Mevduat(Deposit) Nedir?**

Mevduat, bankalara ve benzeri kredi kurumlarına istenildiğinde ya da belli bir vade ya da ihbar süresi sonunda çekilmek üzere yatırılan paraya denir.

### 4.3 Sayısal Değişkenler Arasındaki Korelasyonun İncelenmesi

* __Korelasyon__, numeric tipindeki değişkenlerin birbirleri arasında ilişki olup olmadığını veren istatistiksel bir yöntemdir. 

* __Korelasyon katsayısı (r) nın yorumu;__
    * r<0.2 ise çok zayıf ilişki yada korelasyon yok
    * 0.2-0.4 arasında ise zayıf korelasyon
    * 0.4-0.6 arasında ise orta şiddette korelasyon
    * 0.6-0.8 arasında ise yüksek korelasyon
    * 0.8> ise çok yüksek korelasyon olduğu yorumu yapılır.

**Deposit** yani target değişkeniyle diğer sayısal değişkenler arasındaki ilişkiyi gözlemleyebilmek için 'deposit' değişkenini sayısal bir tipe dönüştürelim.

* `1`:`yes`
* `0`:`no`

#### 4.3.1 Vadeli Mevduat Açmada Görüşme Süresinin Etkisi

* Müşteri ile temas kurma süresinin ortalaması 372 sn. görünmektedir. Bu ortalamanın üzerinde olan müşterilerin vadeli mevduat açma olasılıkları daha yüksek mi?

```python
avgDuration = data['duration'].mean()
print(f"""Ortalamanın üstünde süre ile iletişim kurulan müşterilerin sayısı : {data[data['duration'] > avgDuration]['duration'].count()} 
Ortalamanın altında süre ile iletişim kurulan müşterilerin sayısı : {data[data['duration'] < avgDuration]['duration'].count()}""")
```

```python
data['duration_status'] = np.select([(data['duration']< avgDuration)], ["Ortalamanın Altında"],default="Ortalamanın Üstünde")
```

```python
# satır adı : 'duration_status' - kolon adı : 'deposit' değerlerinden oluşan bir crosstab oluşturalım.
pd.crosstab(data['duration_status'], data['deposit'])
```

```python
# Ortalamanın_altında deposit değeri 0 olanların yüzdesi = 5003/(5003+2314) * 100  :68
durCrosstab = pd.crosstab(data['duration_status'], data['deposit']).apply(lambda r: round(r/r.sum(), 2) * 100, axis=1)

sns.set(rc={'figure.figsize':(12,8)})
sns.set(font_scale=1.2)

ax = durCrosstab.plot(kind='bar', stacked=False, cmap='RdBu')
plt.title("Vadeli Mevduat Açmada Sürenin Etkisi", fontsize=18)
plt.xlabel("Duration Status", fontsize=17);
plt.ylabel("Yüzde (%)", fontsize=17)

for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() *1, p.get_height() * 1))

plt.show()
```

* Müşteri iletişimi için harcanan sürenin(duration) `ortalamasının üstünde` olan müşterilerin `%77'nin Mevduat Hesabı(deposit) açmıştır`. `Ortalamanın altında` süre ile iletişime geçilen müşterilerin sadece `%32'sinin Mevduat Hesabı açmıştır`. <font color = 'green'> **Bu bize, ortalamanın üzerinde görüşülen müşterilerin hedeflenmesinin iyi bir fikir olacağını söylemektedir.**  

```python
data['duration_status'] = np.select(
    [
        data['duration'].between(0, 400, inclusive=True), 
        data['duration'].between(400, 800, inclusive=True),
        data['duration'].between(800, 1200, inclusive=True), 
        data['duration'].between(1200, 1600, inclusive=True)    

    ], 
    [
        '0-400 sn arası', 
        '400-800 sn arası',
        '800-1200 sn arası', 
        '1200-1600 sn arası'
    ], 
    default='1600 sn uzun'
)
```

```python
durationCrosstab = pd.crosstab(data['duration_status'], data['deposit']).apply(lambda r: round(r/r.sum(), 2) * 100, axis=1)

sns.set(rc={'figure.figsize':(12,8)})
sns.set(font_scale=1.2)

ax = durationCrosstab.plot(kind='bar', stacked=False, cmap='RdBu')
plt.title("Vadeli Mevduat Açmada Sürenin Etkisi", fontsize=18)
plt.xlabel("Duration", fontsize=17);
plt.ylabel("Yüzde (%)", fontsize=17)

for p in ax.patches:
    ax.annotate(str(round(p.get_height())), (p.get_x() *1, p.get_height() * 1))

plt.show()
```

* Görüldüğü üzere 400 saniyesinin üstünde görüşmelerde Vadeli Mevduat Açma olasılığının arttığı görülmekte. Yukarıda da belirtilediği üzere <font color = 'green'> **bu bize, ortalamanın üzerinde görüşülen müşterilerin hedeflenmesinin iyi bir fikir olacağını söylemektedir.**

### 4.5 Müşteri Bakiye(Balance) ve Medeni Durumu(Marital) Arasındaki İlişki 

```python
data["marital"].value_counts()
```

```python
sns.set_style('ticks')

fig, ax = plt.subplots(figsize=(15,8))
sns.scatterplot(data=data, x=data.index, y='balance', hue='marital',ax=ax)
plt.ylabel("Balance",fontsize=16)
plt.show()
```

* Nokta dağılımında kesin birşey söylemek mümkün olmamakta. Bunun için değişkenler arasında anlamlılığı görebilmek için `tek yönlü anova` testi uygulayalım. Bunun için hipotezler: 

**H0:** Müşterilerin bakiyeleri ve medeni durumları arasında istatistiksel olarak anlamlı fark bulunmamaktadır. 

**H1:** Müşteri bakiyeleri ve medeni durumları arasında fark bulunmaktadır. 


> **Not:** Test istatistiği olarak `p-value` kullanılacaktır. __P-Value 0,05 den küçük ise H0 ret edilecektir.__

```python
CategoryGroupLists=data.groupby('marital')['balance'].apply(list)
CategoryGroupLists
```

```python
AnovaResults = f_oneway(*CategoryGroupLists)
print('Varyans Analizi sonucunda p-value değeri: ', AnovaResults[1])
```

* Burada p-value değeri 0.019<0.05 olduğundan anlamlılık testi için olan H0 hipotezi ret edilir. Yani müşterilerin evlilik durumu ile banka hesapları arasındaki bir ilişki soz konusudur. 

**Marital Status:** Boşanmanın etkisi `Müşterinin bireysel bakiyesi(balance)` üzerinde önemli bir etkisi vardır.

## 6. Model Seçme

* Bir kaç sınıflandırma algoritması sonuçlarını çapraz doğrulama kullanarak inceleyelim.

```python
dictClassifiers = {
    "Logistic Regression": LogisticRegression(),
    "Linear SVM": SVC(),
    "Nearest Neighbors": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB(),    
    "Gradient Boosting Classifier": GradientBoostingClassifier(),
    "Decision Tree": tree.DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(n_estimators=18),
    "Neural Net": MLPClassifier(alpha=1),
    "XGBoost Classifier": XGBClassifier(eval_metric=['logloss','auc','error'])
}
```

```python
#  Kullanılacak modellerin sayısını değişkende tutmaktadır.
lenClassifiers = len(dictClassifiers.keys())

def topluSiniflandirma(X_train, Y_train, verbose = True):
    
    #(Model sayısı,3) boyutunda 0 değerlerinden oluşan bir dataframe oluşturur.
    dfResults = pd.DataFrame(data=np.zeros(shape=(lenClassifiers,3)), columns = ['classifier_model', 'train_score', 'training_time'])
    count = 0
    for key, classifier in dictClassifiers.items():
        tStart = time.process_time()     #Eğitimden önceki zaman
        classifier.fit(X_train, Y_train) #Eğitim
        tEnd = time.process_time()       #Eğitimden sonraki zaman
        tDiff = tEnd - tStart            #Eğitim başlangıç ve bitiş arasındaki geçen süre
        trainScore = classifier.score(X_train, Y_train)
        dfResults.loc[count,'classifier_model'] = key
        dfResults.loc[count,'train_score'] = trainScore
        dfResults.loc[count,'training_time'] = tDiff
        if verbose:
            print("{c} modelinin eğitimde harcadığı süre : {f:.5f} s".format(c=key, f=tDiff))
        count+=1
    return dfResults
```

```python
df_results = topluSiniflandirma(X, y) #Tüm veri ile capraz doğrulama yapıldı
print('\n',df_results.sort_values(by='train_score', ascending=True))
```

##### <font color = 'blue'> **Overfitting'in Kısa Açıklaması?**

    Aşırı öğrenme (overfitting) , algoritmanın eğitim verisi üzerinden en alt kırılıma kadar çalışıp, sonuçları ezberlemesi ve sadece o veriler üzerinde başarı elde edebilmesidir. Eğitim verisi ile kurduğunuz modeli, test verisi üzerinde çalıştırdığınızda muhtemelen sonuçlar eğitim verisine göre çok düşük çıkacaktır. Modelimizin yapmasını istediğimiz şey, potansiyel bir müşterinin vadeli mevduata abone olup olmayacağını doğru bir şekilde sınıflandırmak için verilerin genel modelini almaktır. Yukarıdaki örneklerde, `Decision Tree`, `Random Forest` ve `XGBoost Classifier` modellerinin bize neredeyse mükemmel doğruluk oranları (%100,%99 ve %95) verdiğinden, modellerin aşırı öğrenmeye yakalanması olasıdır.


<font color = 'blue'> **Overfitting'i nasıl önleyebiliriz?**
    
    Aşırı öğrenmeyi önlemenin en iyi alternatifi çapraz doğrulama kullanmaktır. Eğitim testini almak ve bölmek. Örneğin 3'e bölersek verinin 2/3'ü ya da %66'sı eğitim için, 1/3'ü ise %33'ü veya test için kullanılacak ve test işlemini üç kez gerçekleştirilecektir. Bu algoritma, tüm eğitim ve test setleri boyunca yinelenecektir ve bunun temel amacı, verilerin genel modelini almaktır.

## 7. Hiper-parametre Optimizasyonu

* Optuna ile Gradient Boosting Classifier ait hiper-parametreleri optimize edip X_test, X_train, y_test, y_train değişkenleri ile test edelim

```python
def objective(trial: Trial,X,y) -> float:

    param = {
                "n_estimators" :trial.suggest_int('n_estimators', 0, 1000),
                'max_depth':trial.suggest_int('max_depth', 2, 100),
                'learning_rate':trial.suggest_loguniform('learning_rate',0.005,0.5),
                'subsample':trial.suggest_discrete_uniform('subsample',0.5,1,0.1),
                'min_samples_split':trial.suggest_int('min_samples_split', 0, 1000),
                'min_samples_leaf':trial.suggest_int('min_samples_leaf', 0, 1000)  
#                 'nthread' : -1
            }
    
    model = GradientBoostingClassifier(**param)
    
    return cross_val_score(model, X, y, cv=3,scoring="accuracy").mean()

study = optuna.create_study(direction='maximize',sampler=TPESampler())
study.optimize(lambda trial : objective(trial,X_train,y_train),n_trials= 50)
print('Best trial: score {},\nparams {}'.format(study.best_trial.value,study.best_trial.params))
```

```python
gbcParam = study.best_trial.params.values()
gbcParam
```

```python
gbc = GradientBoostingClassifier(
    n_estimators = list(gbcParam)[0],
    max_depth = list(gbcParam)[1],
    learning_rate = list(gbcParam)[2],
    subsample = list(gbcParam)[3],
    min_samples_split = list(gbcParam)[4],
    min_samples_leaf = list(gbcParam)[5],
    random_state = 42
)
```

```python
gbcModel = gbc.fit(X_train,y_train)
yPred = gbcModel.predict(X_test)
```

* Hiper-parametre optimizasyonu ile skorlarımızı görüntüleyelim. Bunun için ilk olarak Confusion Matrix (Hata Matrisini) oluşturalım.

### 7.1 Metrikler

```python
# Confusion Matrix 
cm = confusion_matrix(y_test, yPred)

group_names = ['True Neg',"False Pos","False Neg","True Pos"]
group_counts = ["{0:0.0f}".format(value) for value in
                cm.flatten()]
group_percentages = ["{0:.2%}".format(value) for value in
                     cm.flatten()/np.sum(cm)]
labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in
          zip(group_names,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(2,2)
sns.heatmap(cm, annot=labels, fmt="", cmap='rocket_r')
plt.show()
```

* True Positive ve True Negative modelin doğru olarak tahminlediği, False Positive ve False Negative ise modelin yanlış olarak tahminlediği alanlardır.
    * `True Positive`: Gelecek verinin sınıfını `positive` (1) olarak tahmin ettik ve bu doğru.
    * `True Negative`: Gelecek verinin sınıfını `positive` (1) olarak tahmin **etmedik** ve bu doğru.
    * `False Positive (Type 1 Error)`: Gelecek verinin sınıfını `positive` (1) olarak tahmin ettik ve bu yanlış .
    * `False Negative (Type 2 Error)`: Gelecek verinin sınıfını `positive` (1) olarak tahmin **etmedik** ve bu yanlış.

**Doğruluk (Accuracy):** Modelde doğru tahmin ettiğimiz alanların toplam veri kümesine oranı ile hesaplanmaktır.
TP+TN/(TP+TN+FN+FP)



```python
print("Modelin Accuracy Score : ",accuracy_score(y_test, yPred))
```

**Kesinlik (Precision):** Positive olarak tahminlediğimiz değerlerin gerçekten kaç adedinin Positive olduğunu göstermektedir. TP / (TP + FP)
    
    Kesinlik değeri özellikle False Positive tahminlemenin maliyeti yüksek olduğu durumlarda çok önemlidir. Örneğin mail kutunuza gelmesi gereken mailleri eğer modeliniz spam olarak işaretlerse (FP) bu durumda almanız gereken önemli mailleri görememiş olur ve sizin için kayıp yaratan bir durumda kalırsınız. Bu durumda Kesinlik değerinin yüksek olması bizim için model seçiminde önemli bir kriterdir.

```python
print("Modelin Precision Score : ",precision_score(y_test, yPred))
```

**Duyarlılık (Recall):** Positive olarak tahmin etmemiz gereken işlemlerin ne kadarını Positive olarak tahmin ettiğimizi gösteren bir metriktir. TP / (TP+FN)

    Duyarlılık değeri de False Negative olarak tahminlemenin maliyetinin yüksek olduğu durumlarda bize yardımcı olacak bir metriktir. Mümkün olduğunca yüksek olması gereklidir. Örneğin Fraud Detection için oluşturduğumuz bir model Fraud olan bir işlemi non-fraud olarak işaretlerse böyle bir durumun sonuçları bir banka için sorun teşkil edecektir.

* ROC farklı sınıflar için bir olasılık eğrisidir. Tipik bir ROC eğrisinde X ekseninde Yanlış Pozitif Oran (FPR) ve Y ekseninde Gerçek Pozitif Orana (TPR) vardır.

* ROC eğrisi kullanmanın yararlarından birkaçı:

    * Farklı modellerin eğrileri genel olarak veya farklı eşikler için doğrudan karşılaştırılabilir.

    * Genellikle, başarılı modeller arsanın sol üst kısmına eğilen eğriler ile temsil edilir.

* AUC: ROC Eğrisi Altındaki Alan

    * Günümüzde en yaygın kullanılan ölçütlerden biri AUC-ROC (Eğri Altındaki Alan — Alıcı Çalışma Karakteristikleri) eğrisidir. AUC, “ROC Eğrisi altındaki alan” anlamına gelir. Bu alanın kapsamı AUC’dir. Kapsanan alan ne kadar büyükse, makine öğrenme modelleri o kadar iyi verilen sınıfları ayırt etmede daha iyidir. AUC için ideal değer 1'dir.
    
`Modelde %85 score alındığı, yani AUC alanı 0.85 büyüklüğünde ve o kadar oranda iyi sınıflandırma yapabilmekte diyebiliriz.`

## 8. Sonuçlar

1) **Pazarlama Hareketi Ayları:** Pazarlama etkinliğinin en yüksek olduğu ayın **Mayıs** olduğunu gördük. Ancak bu, potansiyel müşterilerin vadeli mevduat tekliflerini reddetme eğiliminde olduğu aydır (reddetme oran: %33). Bir sonraki pazarlama kampanyası için, bankanın pazarlama kampanyasına **Mart, Eylül, Ekim ve Aralık** aylarında odaklanması akıllıca olacaktır. (Aralık, pazarlama etkinliğinin en düşük olduğu ay olduğu için dikkate alınmalıdır. Aralık ayının en düşük olmasının bir nedeni olabilir.)<br><br>
2) **Yaş Kategorisi:** Bankanın bir sonraki pazarlama kampanyasında, `25 veya daha küçük ve 60 veya daha büyük` yaşta olan potansiyel müşterileri hedeflemelidir. 25 yaş ve altında olan müşterilerin vadeli mevduata abone olma şansı `%71` iken, 60 yaş ve üstündekilerin vadeli mevduata abone olma şansı `%82'dir`. Bir sonraki kampanya için bankanın bu iki kategoriyi ele alması ve dolayısıyla daha fazla vadeli mevduat talebi olasılığını artırması harika olacaktır. <br><br>
3) **Meslek:** Öğrenci veya emekli olan potansiyel müşterilerin vadeli mevduata başvurma olasılıkları en yüksekti. Emekli bireyler, faiz ödemeleri yoluyla bir miktar nakit kazanmak için daha fazla vadeli mevduata sahip olma eğilimindedir. Vadeli mevduatlar, kişinin (bu durumda emekli kişinin) kişi ve finans kurumu arasında kararlaştırılan belirli bir tarihe kadar bankadan nakit çekmemeyi kabul ettiği kısa vadeli kredilerdir. Bu süreden sonra kişi sermayesini geri alır ve krediye yaptığı faizi alır. Emekli bireyler, nakit parasını büyük ölçüde harcamama eğilimindedir, bu nedenle paralarını finansal kuruma borç vererek işe koyma olasılıkları daha yüksektir. <br><br>
4) **Konut Kredileri(Loan) ve Bakiyeler(Balance):** Düşük bakiyeli ve bakiyesiz kategorisindeki potansiyel müşterilerin, ortalama ve yüksek bakiye kategorisindeki kişilere göre konut kredisi alma olasılığı daha yüksekti. Ev kredisi almak ne anlama geliyor? Bu, potansiyel müşterinin konut kredisini geri ödemek için finansal tavizleri olduğu ve bu nedenle vadeli mevduat hesabına açabileceği nakdi olmadığı anlamına gelir. Ancak ortalama ve yüksek bakiyelerdeki potansiyel müşterilerin konut kredisi kullanma olasılıklarının daha düşük ve dolayısıyla vadeli mevduat açma olasılıklarının daha yüksek olduğunu görüyoruz. Son olarak, bir sonraki pazarlama kampanyası, vadeli mevduata abone olma olasılığını artırmak için ortalama ve yüksek bakiyeli bireylere odaklanmalıdır. <br><br>
5) **Kampanya Çağrıları:** Yeni potansiyel müşteriler edinmede zamandan ve emekten tasarruf etmek için aynı potansiyel müşteriye en fazla 3 çağrının uygulanması gerektiğini belirten bir politika uygulanmalıdır. Unutmayın, aynı potansiyel müşteriyi ne kadar çok ararsak, o büyük olasılıkla vadeli mevduat açmayı reddedecektir. <br><br>
6) **Aramalar Sırasında Anket Oluşturun:** Görüşme süresi, potansiyel bir müşterinin vadeli mevduat açıp açmayacağı ile en olumlu şekilde ilişkili olan özellik olduğundan, görüşme sırasında potansiyel müşteriler için ilginç bir anket sağlayarak görüşmenin uzunluğu artabilir. Elbette bu, potansiyel müşterinin vadeli mevduat kabul edeceğini garanti etmez! Bununla birlikte, potansiyel müşterinin katılım düzeyini artıracak ve vadeli mevduata bağlanma olasılığını artıracak bir strateji uygulayarak hiçbir şey kaybetmeyiz ve bu nedenle bankanın gerçekleştireceği bir sonraki pazarlama kampanyası için etkinlikte bir artış sağlarız. . <br><br>
7) **Vadesi daha yüksek (372 üzeri) bireyleri hedefleyin:** Süre olarak ortalamanın üzerinde görüşülen hedef kitleyi hedefleyin. Bu grubun vadeli mevduat hesabı açma olasılığı %77 ile oldukça yüksek. Bir sonraki pazarlama kampanyasının başarı oranının oldukça başarılı olmasını sağlar.<br><br>

Tüm bu stratejileri birleştirerek ve bir sonraki kampanyanın hitap etmesi gereken pazar kitlesini basitleştirerek, bankanın bir sonraki pazarlama kampanyasının mevcut olandan daha etkili olması muhtemeldir.
